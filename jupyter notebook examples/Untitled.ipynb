```python
import ast
import numpy as np

#  eval 
SAFE_GLOBALS = {'np': np}

def python_to_str(value) -> str:
    """
     Python  eval 
      - None           -> ''
      - /    -> repr(value)
      - list/dict/tuple -> repr(value)
      - 1D ndarray  -> np.linspace(...)  np.arange(...)
      -  ndarray   -> np.array([...])
    """
    # 1) None
    if value is None:
        return ''
    # 2) ndarray 
    if isinstance(value, np.ndarray):
        # >1
        if value.ndim == 1 and value.size > 1:
            diffs = np.diff(value)
            #  -> linspace
            if np.allclose(diffs, diffs[0], rtol=1e-6, atol=1e-12):
                start = value[0]
                stop  = value[-1]
                n     = value.size
                return f"np.linspace({start!r}, {stop!r}, {n})"
            #  arange
            step = diffs[0]
            expected = np.arange(start, start + step*value.size, step)
            if expected.shape == value.shape and np.allclose(expected, value):
                end = start + step*value.size
                return f"np.arange({start!r}, {end!r}, {step!r})"
        #  ndarray 
        return "np.array(" + repr(value.tolist()) + ")"
    # 3)  repr
    text = repr(value)
    # 
    text = text.replace('\n', '')
    return text

def str_to_python(s: str):
    """
     QLineEdit  Python 
      1) ast.literal_eval ///tuple/'None'
      2) eval(s, SAFE_GLOBALS)  np.array/... 
      3) 
    """
    s = s.strip()
    if not s:
        return None
    try:
        #  literal
        return ast.literal_eval(s)
    except Exception:
        try:
            #  np.* 
            return eval(s, SAFE_GLOBALS, {})
        except Exception:
            return s
```

```python
def float2str(value, digits=None, length=None):
    """
    Convert float to string taking the first `digits` numeric characters (including zeros), excluding sign and decimal point.

    Args:
        value (float): number to format.
        digits (int, optional): total numeric characters to include.
        length (int, optional): total output length including sign and decimal point.

    Returns:
        str: formatted string.

    Examples:
        >>> float2str(123.456, digits=4)
        '123.4'
        >>> float2str(0.001234, digits=3)
        '0.00'
        >>> float2str(-9.8765, digits=2)
        '-9.8'
        >>> float2str(3.14, length=5)
        '3.14 '
    """
    # Prepare full fixed-point string with extra precision
    s_full = f"{value:.{max(digits or 0,0)+10}f}" if digits else str(value)
    result = ''
    num_count = 0
    # Preserve sign
    if s_full.startswith('-'):
        result += '-'
        s_full = s_full[1:]
    # Iterate to collect numeric chars
    for ch in s_full:
        if ch.isdigit():
            if digits and num_count >= digits:
                break
            result += ch
            num_count += 1
        elif ch == '.':
            result += '.'
        # stop if decimal portion done and no more digits needed
    # Remove trailing '.'
    if result.endswith('.'):
        result = result[:-1]
    # Enforce total length
    if length:
        result = result[:length].ljust(length)
    return result


def float2str_eng(value, length):
    """
    Convert a float to either engineering or plain form, choosing shortest full-precision
    representation if under max length; otherwise choose most precise form.

    Args:
        value (float): number to format.
        length (int): maximum output characters (including sign, digits, decimal point, exponent).

    Returns:
        str: chosen representation.

    Behavior:
      1. Build full-precision plain and engineering strings.
      2. If either fits within `length`, return the shorter that fits.
      3. If neither fits, return the shorter full-precision string regardless of length.

    Examples:
        >>> float2str_eng(1234567, 10)
        '1.234567e6'  # fits lengths, choose shorter
        >>> float2str_eng(1234567, 5)
        '1.234567e6'  # full precision longer than length but return most precise
    """
    # Generate full precision plain: 17 significant digits
    plain_full = f"{value:.17g}"
    # Engineering full precision
    if value == 0:
        eng_full = '0'
    else:
        exp = int(np.floor(np.log10(abs(value))))
        eng_exp = exp - (exp % 3)
        mant = value / (10 ** eng_exp)
        eng_full = f"{mant:.17g}e{eng_exp}"
    # Collect candidates
    candidates = []
    # Consider plain if fits
    if len(plain_full) <= length:
        candidates.append(plain_full)
    # Consider engineering if fits
    if len(eng_full) <= length:
        candidates.append(eng_full)
    if candidates:
        # Return shortest among fitting representations
        return min(candidates, key=len)
    # If none fits, return shortest full-precision form
    return plain_full if len(plain_full) <= len(eng_full) else eng_full
```
